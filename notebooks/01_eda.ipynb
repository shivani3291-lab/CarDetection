# 01 — Exploratory Data Analysis

Stanford Cars dataset: class distribution, bbox stats, sample grid.

```python
import json
from collections import Counter
from pathlib import Path

import matplotlib.pyplot as plt
from PIL import Image

ROOT = Path('..').resolve()
manifest = json.loads((ROOT / 'data/processed/manifest.json').read_text())
print(f'Samples: {len(manifest)}')
```

```python
train = [r for r in manifest if r['split'] == 'train']
counts = Counter(r['class_name'] for r in train)
fig, ax = plt.subplots(figsize=(14, 4))
ax.bar(range(len(counts)), list(counts.values()), color='#7F77DD')
ax.set_title('Class distribution (train)')
plt.show()
```

```python
widths, heights = [], []
for r in train[:500]:
    x1, y1, x2, y2 = r['bbox']
    widths.append(x2 - x1)
    heights.append(y2 - y1)

fig, axes = plt.subplots(1, 2, figsize=(10, 3))
axes[0].hist(widths, bins=30, color='#7F77DD')
axes[0].set_title('BBox width')
axes[1].hist(heights, bins=30, color='#7F77DD')
axes[1].set_title('BBox height')
plt.tight_layout()
plt.show()
```

```python
from src.serve.viz import draw_boxes

fig, axes = plt.subplots(5, 5, figsize=(12, 12))
for ax, r in zip(axes.flat, train[:25]):
    img = Image.open(r['image_path']).convert('RGB')
    ann = draw_boxes(img, [{'class': r['class_name'], 'score': 1.0, 'bbox': r['bbox']}])
    ax.imshow(ann)
    ax.axis('off')
plt.suptitle('Sample grid with ground-truth bboxes')
plt.tight_layout()
plt.show()
```
